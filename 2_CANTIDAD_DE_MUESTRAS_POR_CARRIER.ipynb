```python
import os
import pandas as pd

# Directorio de las carpetas
ruta_muestras = r'D:\Python\TESIS\CLUSTERS_C4_CALC_K\train'

# Inicializamos listas para almacenar los nombres de las carpetas y el número de archivos
carpetas = []
num_muestras = []

# Recorremos las carpetas en la ruta de MUESTRAS
for carpeta in os.listdir(ruta_muestras):
    ruta_carpeta = os.path.join(ruta_muestras, carpeta)
    if os.path.isdir(ruta_carpeta):
        # Contamos el número de archivos dentro de cada carpeta
        num_archivos = len(os.listdir(ruta_carpeta))
        carpetas.append(carpeta)
        num_muestras.append(num_archivos)

# Crear un DataFrame con la información
df_muestras = pd.DataFrame({
    'MUESTRAS POR TIPO': carpetas,
    'Cant. de muestras': num_muestras
})

# Calcular el total de muestras
total_muestras = df_muestras['Cant. de muestras'].sum()

# Agregar la columna con el porcentaje
df_muestras['Porcentaje (%)'] = (df_muestras['Cant. de muestras'] / total_muestras) * 100

# Redondear el porcentaje a 1 decimal
df_muestras['Porcentaje (%)'] = df_muestras['Porcentaje (%)'].round(3)

# Agregar una fila con el total
total_row = pd.DataFrame({
    'MUESTRAS POR TIPO': ['Total'],
    'Cant. de muestras': [df_muestras['Cant. de muestras'].sum()],
    'Porcentaje (%)': [df_muestras['Porcentaje (%)'].sum()]
})

# Concatenar el DataFrame original con la fila de totales
df_muestras = pd.concat([df_muestras, total_row], ignore_index=True)

# Mostrar la tabla
display(df_muestras)
```
