```python
import os
from PIL import Image
```

```python
# Ruta de la carpeta principal de MUESTRAS
base_folder = r'D:\Python\TESIS\MUESTRAS_V2'
# Ruta de destino para las imgenes con fondo blanco
output_base_folder = r'D:\Python\TESIS\MUESTRAS_CENTRADAS_V2'

# Crear una funcin para centrar la imagen en un fondo blanco de 170x170
def center_image_on_background(image, background_size=(170, 170), color=(255, 255, 255)):
    # Crear una nueva imagen de fondo
    background = Image.new("RGB", background_size, color)
    
    # Calcular las coordenadas para centrar la imagen
    x_offset = (background_size[0] - image.size[0]) // 2
    y_offset = (background_size[1] - image.size[1]) // 2
    
    # Pegar la imagen en el fondo
    background.paste(image, (x_offset, y_offset))
    
    return background

# Recorrer todas las carpetas y subcarpetas
for root, dirs, files in os.walk(base_folder):
    for file in files:
        if file.endswith(".png") or file.endswith(".jpg"):  # Filtrar por imgenes
            file_path = os.path.join(root, file)
            
            # Abrir la imagen original
            img = Image.open(file_path)
            
            # Centrar la imagen en el fondo blanco de 170x170
            centered_img = center_image_on_background(img)
            
            # Crear la estructura de carpetas de salida, similar a MUESTRAS
            relative_path = os.path.relpath(root, base_folder)
            output_dir = os.path.join(output_base_folder, relative_path)
            if not os.path.exists(output_dir):
                os.makedirs(output_dir)
            
            # Guardar la nueva imagen en la carpeta de salida
            new_file_path = os.path.join(output_dir, f"centered_{file}")
            centered_img.save(new_file_path)
            
            print(f"Guardando {new_file_path}.")

print("Proceso completado.")
```
